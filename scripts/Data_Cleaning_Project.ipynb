```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


df = pd.read_csv('AB_NYC_2019.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
#showing the null values on heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(df.isnull())
```

```python
df['last_review'] = pd.to_datetime(df['last_review'])

# Display rows where both 'last_review' and 'reviews_per_month' are null
missing_reviews = df[df['last_review'].isnull() & df['reviews_per_month'].isnull()]
print(f"Number of rows where both 'last_review' and 'reviews_per_month' are null: {len(missing_reviews)}")
display(missing_reviews.head())
```

This confirms that the missing values in `last_review` and `reviews_per_month` occur in the same rows. This suggests these listings have never received a review. We can also visualize the distribution of `reviews_per_month` for non-missing values and see how `last_review` dates are distributed.

```python
# Distribution of 'reviews_per_month' for non-missing values
plt.figure(figsize=(5, 3))
sns.histplot(df['reviews_per_month'].dropna(), bins=50, kde=True)
plt.title('Distribution of Reviews Per Month (excluding NaN)')
plt.xlabel('Reviews Per Month')
plt.ylabel('Count')
plt.show()
```

```python
# Distribution of 'last_review' by year
plt.figure(figsize=(5, 3))
sns.histplot(df['last_review'].dropna().dt.year, bins=range(2010, 2020), discrete=True)
plt.title('Distribution of Last Review Year (excluding NaN)')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(range(2010, 2020))
plt.show()
```

```python
# Dropping the first two rows with missing values since their missing values are just 0.04% of the dataset
df = df.dropna(subset=['name', 'host_name'])
```

```python
df.info()
```

```python
#Filling the null values with 0 since there's no review for it
df['reviews_per_month'].fillna(0, inplace=True)
df['last_review'].fillna(pd.NaT, inplace=True)  #(means no review yet)
```

*   **Missing Data Handling**: Missing values were identified and addressed systematically.

    *   Missing values in 'reviews_per_month' were imputed with `0`, logically signifying no reviews.
    *   Missing values in 'last_review' were imputed with `pd.NaT`, indicating the absence of a last review date, consistent with listings having no reviews.
*   **Data Standardization**: All column names were found to be consistently in lowercase, requiring no further standardization efforts.
